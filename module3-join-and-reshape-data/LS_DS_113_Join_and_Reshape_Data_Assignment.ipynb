## Join Data Practice

These are the top 10 most frequently ordered products. How many times was each ordered? 

1. Banana
2. Bag of Organic Bananas
3. Organic Strawberries
4. Organic Baby Spinach 
5. Organic Hass Avocado
6. Organic Avocado
7. Large Lemon 
8. Strawberries
9. Limes 
10. Organic Whole Milk

First, write down which columns you need and which dataframes have them.

Next, merge these into a single dataframe.

Then, use pandas functions from the previous lesson to get the counts of the top 10 most frequently ordered products.

```python
##### YOUR CODE HERE #####
import pandas as pd 
# orders.csv
orders = pd.read_csv('orders.csv')
print(orders.shape)
orders.head()
# need to filter these to only have top 10 products
```

```python
# read in order_products CSVs
order_products = pd.concat([pd.read_csv('order_products__prior.csv'), 
                            pd.read_csv('order_products__train.csv')])
order_products.head()
# can match these to orders using order_id
```

```python
# read in products.csv and filter down to just products we want
products = pd.read_csv('products.csv')
top_ten_products = ['Banana',
                    'Bag of Organic Bananas',
                    'Organic Strawberries',
                    'Organic Baby Spinach',
                    'Organic Hass Avocado',
                    'Organic Avocado',
                    'Large Lemon',
                    'Strawberries',
                    'Limes',
                    'Organic Whole Milk']
top_ten_products = products['product_name'].isin(top_ten_products)
top_ten_products = products[top_ten_products]
top_ten_products.head(10)
```

```python
# merge top_ten_products with order_products
order_products_merged_with_products = pd.merge(
    order_products[['order_id', 'product_id']], 
    top_ten_products[['product_id', 'product_name']])
print(order_products_merged_with_products.shape)
order_products_merged_with_products.head()
```

```python
# merge order_products_merged_with_products with orders
final = pd.merge(orders['order_id'], order_products_merged_with_products, on='order_id', how='inner')
print(final.shape)
final.head()
```

```python
final['product_name'].value_counts()
```

```python
import pandas as pd
import numpy as np

table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T.reset_index()

table2
```

```python
# table2 -> tidy
tidy = table2.melt(
    id_vars='index', 
    value_vars=['John Smith', 'Jane Doe', 'Mary Johnson'])
tidy
```

```python
# tidy -> table2
back_to_table2 = tidy.pivot_table(index='index', columns='variable', values='value')
back_to_table2 
```

## Join Data Stretch Challenge

The [Instacart blog post](https://tech.instacart.com/3-million-instacart-orders-open-sourced-d40d29ead6f2) has a visualization of "**Popular products** purchased earliest in the day (green) and latest in the day (red)." 

The post says,

> "We can also see the time of day that users purchase specific products.

> Healthier snacks and staples tend to be purchased earlier in the day, whereas ice cream (especially Half Baked and The Tonight Dough) are far more popular when customers are ordering in the evening.

> **In fact, of the top 25 latest ordered products, the first 24 are ice cream! The last one, of course, is a frozen pizza.**"

Your challenge is to reproduce the list of the top 25 latest ordered popular products.

We'll define "popular products" as products with more than 2,900 orders.



```python
##### YOUR CODE HERE #####
import pandas as pd 
orders = pd.read_csv('orders.csv')
order_products = pd.concat([pd.read_csv('order_products__prior.csv'), 
                            pd.read_csv('order_products__train.csv')])
products = pd.read_csv('products.csv')
```

```python
merged = pd.merge(pd.merge(orders, order_products, on='order_id', how='inner'), 
                  products, on='product_id', how='inner')
merged.head()
```

```python
merged = merged.drop(['user_id', 'eval_set', 'order_number', 'order_dow', 
                     'days_since_prior_order', 'add_to_cart_order', 'reordered',
                     'aisle_id', 'department_id'], axis=1)
merged.head()
```

```python
# condition = (merged['product_name'].value_counts() > 2900)
new_merged = merged[merged['product_name'].isin(
    merged['product_name'].value_counts()[merged['product_name'].value_counts()>2900].index)]
new_merged['product_name'].value_counts()
```

```python
pd.crosstab(new_merged['order_hour_of_day'], new_merged['product_name'], normalize=True)
```
